```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os


# 폰트 설치 및 시스템 업데이트
!apt-get install fonts-nanum -qq > /dev/null
!apt-get install fontconfig -qq > /dev/null
!fc-cache -fv > /dev/null

# 폰트 경로 직접 지정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'

# 폰트 매니저에 명시적으로 등록
fontprop = fm.FontProperties(fname=font_path, size=12)
fm.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False


# 함수 정의
def f1(x):
    return 2*x + 3

def f2(x):
    return x**2 - 6*x + 8
```

```python

# (1) 꼭짓점 계산
a, b, c = 1, -6, 8
vertex_x = -b / (2*a)
vertex_y = f2(vertex_x)
print(f"f2(x)의 꼭짓점: ({vertex_x}, {vertex_y})")
```

```python

# (2) x절편 계산
# f1(x): 2x + 3 = 0
x1_intercept = -3/2
print(f"f1(x)의 x절편: {x1_intercept}")

# f2(x): x^2 - 6x + 8 = 0
# 이차방정식 근의 공식 사용
x2_intercepts = np.roots([1, -6, 8])
print(f"f2(x)의 x절편: {x2_intercepts}")

# (3) y절편 계산
y1_intercept = f1(0)
y2_intercept = f2(0)
print(f"f1(x)의 y절편: {y1_intercept}")
print(f"f2(x)의 y절편: {y2_intercept}")
```

```python
# 그래프 그리기
x = np.linspace(-2, 8, 100)
plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label='f1(x) = 2x + 3')
plt.plot(x, f2(x), label='f2(x) = x^2 - 6x + 8')

# 꼭짓점, x절편, y절편 표시
plt.plot(vertex_x, vertex_y, 'ro', label='f2(x) 꼭짓점')
plt.plot(x1_intercept, 0, 'go', label='f1(x) x절편')
plt.plot(x2_intercepts, [0, 0], 'bo', label='f2(x) x절편')
plt.plot(0, y1_intercept, 'mo', label='f1(x) y절편')
plt.plot(0, y2_intercept, 'co', label='f2(x) y절편')

plt.title('함수 그래프 분석')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.show()
```
