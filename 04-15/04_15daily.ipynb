## 문제 1. 지도학습에서 레이블이 필요한 이유
다음 중 **지도학습에서 "정답 레이블(label)"이 필요한 이유**로 가장 적절한 설명은?

1. 데이터를 시각적으로 보기 쉽게 정렬하기 위해  
2. 입력 데이터의 차원을 줄이기 위해  
3. 모델이 입력과 출력 간의 관계를 학습할 수 있도록 하기 위해  
4. 분류(class) 이름을 예쁘게 출력하기 위해  

👉 정답: (3 )  
👉 간단한 이유: 입력과 출력간의 관계를 학습하기위해 정답레이블이 필요하다
